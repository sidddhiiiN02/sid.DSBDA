```python
df.boxplot(column=['ssc_p','hsc_p','degree_p','mba_p'])
plt.title("Boxplot of Academic Scores")
plt.show()
```

```python
z = np.abs(stats.zscore(df['mba_p']))
df[z > 3]
```

```python
Q1 = df['hsc_p'].quantile(0.25)
Q3 = df['hsc_p'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print("Lower Bound:", lower)
print("Upper Bound:", upper)
```

```python
df[(df['hsc_p'] < lower) | (df['hsc_p'] > upper)]
```

```python
median = df['hsc_p'].median()
df['hsc_p'] = np.where((df['hsc_p'] < lower) | (df['hsc_p'] > upper),median,df['hsc_p'])
```

```python
df['salary'].hist()
plt.title("Salary Before Log")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
df['salary'].skew()
```

```python
df['log_salary'] = np.log10(df['salary'])
df['log_salary'].hist()
plt.title("Salary After Log")
plt.xlabel("Log Salary")
plt.ylabel("Frequency")
plt.show()
df['log_salary'].skew()
```

```python
plt.scatter(df['mba_p'], df['salary'])
plt.xlabel("MBA Percentage")
plt.ylabel("Salary")
plt.title("MBA vs Salary")
plt.show()
```

```python
mad = (df['salary'] - df['salary'].mean()).abs().mean()
print("Mean Absolute Deviation:", mad)
```

```python
variance = df['salary'].var()
print("Variance:", variance)
```

```python
std_dev = df['salary'].std()
print("Standard Deviation:", std_dev)
```

```python
data_range = df['salary'].max() - df['salary'].min()
print("Range:", data_range)
```

```python
def dgraph(df, variable):
    plt.figure(figsize=(15,6))  
    plt.subplot(1, 2, 1)
    stats.probplot(df[variable], dist='norm', plot=plt)
    plt.title("Q-Q Plot")
    plt.show()
```

```python
dgraph(df, 'salary')
```
