```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data =pd.read_csv("C:/Users/Siddhi/Downloads/boston prac4 dataset/boston.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data= data.drop(["CHAS"],axis=1)
```

```python
data = data.drop(["ZN"],axis=1)
```

```python
data
```

```python
X = data.drop("MEDV", axis=1)
y = data["MEDV"]
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split( X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
ytrain_pred = model.predict(xtrain)
ytest_pred = model.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print("Training MSE:", mean_squared_error(ytrain, ytrain_pred))
print("Testing MSE:", mean_squared_error(ytest, ytest_pred))
print("Training R2:", r2_score(ytrain, ytrain_pred))
print("Testing R2:", r2_score(ytest, ytest_pred))
```

```python
plt.scatter(ytrain ,ytrain_pred,c='pink',marker='o',label='Training data')
plt.scatter(ytest,ytest_pred ,c='grey',marker='s',label='Test data')
plt.xlabel('True values')
plt.ylabel('Predicted')
plt.title("True value vs Predicted value")
plt.legend(loc='upper left')
plt.plot()
plt.show() 
```
