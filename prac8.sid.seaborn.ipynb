```python
pt=[23,26,34,38,43,56]
```

```python
plt.plot(pt)
plt.show()
```

```python
plt.plot(pt)
plt.xlabel("sample point")
plt.ylabel("values")
plt.title("point vs value")
```

```python
pt1=[78,34,56,67,27,87,45,33]
```

```python
plt.plot(pt)
plt.plot(pt1)
```

```python
plt.plot(pt , c='b' , marker='o' , mec='r' )
plt.plot(pt1 , c='r', marker='*',mec='b')
plt.xlabel("sample point")
plt.ylabel("values")
plt.title("point vs value")
```

```python
plt.plot(pt , c='b' , marker='o' , mec='r', ls=':')
plt.plot(pt1 , c='r', marker='*',mec='b', ls=':')
plt.xlabel("sample point")
plt.ylabel("values")
plt.title("point vs value")
```

```python
plt.plot(pt , c='b' , marker='o' , mec='r', dash_capstyle ='round')
plt.plot(pt1 , c='r', marker='*',mec='b', ls=':')
plt.xlabel("sample point")
plt.ylabel("values")
plt.title("point vs value")
```

```python
plt.plot(pt , c='b' , marker='o' , mec='r', dash_capstyle ='round')
plt.plot(pt1 , c='r', marker='*',mec='b', ls=':')
plt.xlabel("sample point")
plt.ylabel("values")
plt.title("point vs value")
plt.legend(['cns','dbms'])
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=sns.load_dataset('titanic')
```

```python
dataset.head()
```

1.DISTRIBUTION :
Distplot

```python
sns.distplot(x = dataset['age'],bins = 10)
```

```python
sns.distplot(dataset['age'], bins = 10,kde=False)
```

```python
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind ='scatter')
```

Joint Plot:

```python
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind = 'hex')
```

Rug Plot

```python
sns.rugplot(dataset['fare'])
```

2.CATEGORICAL :
Bar Plot

```python
sns.barplot(x='sex', y='age', data=dataset)
```

```python
sns.barplot(x='sex', y='age', data=dataset, estimator=np.std)
```

Count Plot

```python
sns.countplot(x='sex', data=dataset)
```

Box Plot

```python
sns.boxplot(x='sex', y='age', data=dataset)
```

```python
sns.boxplot(x='sex', y='age', data=dataset, hue="survived")
```

Violin Plot

```python
sns.violinplot(x='sex', y='age', data=dataset)
```

```python
sns.violinplot(x='sex', y='age', data=dataset, hue='survived')
```

3.ADVANCED PLOTS:
Strip Plots:

```python
sns.stripplot(x='sex', y='age', data=dataset, jitter=False)
```

```python
sns.stripplot(x='sex', y='age', data=dataset, jitter=True)
```

```python
sns.stripplot(x='sex', y='age', data=dataset, jitter=True, hue='survived')
```

Swarm Plot:

```python
sns.swarmplot(x='sex', y='age', data=dataset)
```

```python
sns.swarmplot(x='sex', y='age', data=dataset, hue='survived')
```

4.MATRIX PLOTS:
Heat map:

```python
corr=dataset.corr(numeric_only=True)
sns.heatmap(corr)
```

```python
corr = dataset.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```
