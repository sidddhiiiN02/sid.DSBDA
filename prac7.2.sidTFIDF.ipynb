```python
import nltk
import math
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

Sample document

```python
documentA = "Jupiter is the largest planet"
documentB = "Mars is the fourth planet from the Sun"
```

```python
documents=[documentA, documentB]
```

```python
stop_words=set(stopwords.words('english'))
```

1.Preprocessing

```python
processed_docs = []

for doc in documents:
    doc = doc.lower()
    words = word_tokenize(doc)
    filtered_words = [w for w in words if w not in stop_words]
    processed_docs.append(filtered_words)

print("Processed Documents:", processed_docs)

        
```

2.Vocabulary

```python
vocab = set()
for doc in processed_docs:
    vocab.update(doc)

vocab = list(vocab)
```

3.Term Frequency

```python
def compute_tf(doc):
    tf_dict = {}
    doc_len = len(doc)
    for word in vocab:
        tf_dict[word] = doc.count(word) / doc_len
    return tf_dict

tf_docs = [compute_tf(doc) for doc in processed_docs]
```

4.Inverse Document Frequency

```python
def compute_idf(docs):
    N = len(docs)
    idf_dict = {}

    for word in vocab:
        count = sum(1 for doc in docs if word in doc)
        idf_dict[word] = math.log(N / (count + 1)) + 1  # smoothing

    return idf_dict

idf = compute_idf(processed_docs)
```

5.TF-IDF

```python
def compute_tfidf(tf, idf):
    tfidf = {}
    for word in vocab:
        tfidf[word] = tf[word] * idf[word]
    return tfidf

tfidf_docs = [compute_tfidf(tf, idf) for tf in tf_docs]
```

#OUTPUT:

```python
print("\nVocabulary:", vocab)
```

```python
print("\nTF for each document:")
for i, tf in enumerate(tf_docs):
    print(f"Document {i+1}:", tf)
```

```python
print("\nIDF values:")
print(idf)
```

```python
print("\nTF-IDF for each document:")
for i, tfidf in enumerate(tfidf_docs):
    print(f"Document {i+1}:", tfidf)
```
